```python
df = pd.read_csv('data/fish.csv')[['Width', 'Weight']].rename(columns={
    'Width': 'width (cm)',
    'Weight': 'weight (g)'
})

fig, ax = plt.subplots(1, 1, figsize=(4, 4))
plot_data(ax, df, 'width (cm)', 'weight (g)')
```
